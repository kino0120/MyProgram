```python
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris()
df_feature_iris = pd.DataFrame(iris['data'],columns=['sepal_length','sepal_width'
                                                    ,'petal_length','petal_width'])
df_feature_iris.head(5)
```

```python
df_species_iris = pd.DataFrame(iris['target'],columns=['species'])
df_species_iris.head(5)
```

```python
df_iris = pd.concat([df_feature_iris,df_species_iris],axis=1)
df_iris.describe()
```
